```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
```

```python
# Load the dataset
data = load_breast_cancer()
X = data.data
y = data.target

# Create a DataFrame for easier data manipulation
df = pd.DataFrame(X, columns=data.feature_names)
df['target'] = y
```

```python
# Selecting only 'mean radius' and 'mean texture'
X = df[['mean radius', 'mean texture']]

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Standardizing the data to improve Naive Bayes performance
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Create and train the Gaussian Naive Bayes classifier
gnb = GaussianNB()
gnb.fit(X_train, y_train)

# Prediction
y_pred = gnb.predict(X_test)

# Print accuracy and confusion matrix
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
# Create a mesh to plot in
x_min, x_max = X_train[:, 0].min() - 1, X_train[:, 0].max() + 1
y_min, y_max = X_train[:, 1].min() - 1, X_train[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))

# Plot decision boundaries
Z = gnb.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha=0.4, cmap='coolwarm')
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, s=20, edgecolor='k', cmap='coolwarm')
plt.title('Naive Bayes Classification)')
plt.xlabel('Mean Radius)')
plt.ylabel('Mean Texture')
plt.colorbar()
plt.show()
```

```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['diagnosis'] = data.target

# Display basic information
print(df.head())
print(df.describe())

# Extract mean radius and mean texture for plotting
malignant = df[df['diagnosis'] == 0]  # Assuming '0' is malignant
benign = df[df['diagnosis'] == 1]  # Assuming '1' is benign

plt.figure(figsize=(10, 6))
plt.scatter(malignant['mean radius'], malignant['mean texture'], color='red', label='Malignant', alpha=0.5)
plt.scatter(benign['mean radius'], benign['mean texture'], color='blue', label='Benign', alpha=0.5)
plt.title('Scatter Plot of Mean Radius and Mean Texture')
plt.xlabel('Mean Radius')
plt.ylabel('Mean Texture')
plt.legend()
plt.show()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Extracting the features and labels
X = df[['mean radius', 'mean texture']].values
y = df['diagnosis'].values

# Standardize features (important for neural network convergence)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Reshaping data to fit the CNN input shape (n_samples, height, width, channels)
X_reshaped = X_scaled.reshape(-1, 1, 2, 1)  # Here, we have 1 row, 2 columns, and 1 channel

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_reshaped, y, test_size=0.2, random_state=42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, Flatten, MaxPooling2D

# Build the CNN model
model = Sequential()
model.add(Conv2D(16, (1, 2), activation='relu', input_shape=(1, 2, 1)))
model.add(MaxPooling2D((1, 1)))
model.add(Flatten())
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test))
```

```python
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Load data
data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['diagnosis'] = data.target

# Extracting features and labels
X = df[['mean radius', 'mean texture']].values
y = df['diagnosis'].values

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_reshaped = X_scaled.reshape(-1, 1, 2, 1)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_reshaped, y, test_size=0.2, random_state=42)

# Define CNN model
model = Sequential([
    Conv2D(16, (1, 2), activation='relu', input_shape=(1, 2, 1)),
    MaxPooling2D((1, 1)),
    Flatten(),
    Dense(10, activation='relu'),
    Dense(1, activation='sigmoid')
])

# Compile and train the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=10, batch_size=10, verbose=0)

# Predictions for visualization
y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5).astype(int)
```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Load dataset
data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['diagnosis'] = data.target

# Features and Labels
X = df.drop('diagnosis', axis=1)
y = df['diagnosis']

# Data Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Define the model
model = Sequential([
    Dense(16, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(16, activation='relu'),
    Dense(1, activation='sigmoid')
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=50, batch_size=10, verbose=0)
```

```python
# Predictions
y_pred = model.predict(X_test)
y_pred_classes = (y_pred > 0.5).astype(int)
```

```python
# Compute confusion matrix
cm = confusion_matrix(y_test, y_pred_classes)

# Plot confusion matrix
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt="d", cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python
# Classification report for precision, recall, f1-score and accuracy
cr = classification_report(y_test, y_pred_classes, target_names=['Benign', 'Malignant'])
print(cr)
```
